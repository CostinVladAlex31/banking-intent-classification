```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

import torch
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding
)
from datasets import Dataset

print(f'PyTorch: {torch.__version__}')
print(f'GPU: {torch.cuda.get_device_name(0)}')
print('✅ Totul importat cu succes!')
```

```python
train_df  = pd.read_csv('train.csv')
test_df   = pd.read_csv('test.csv')
sample_df = pd.read_csv('sample_submission.csv')

print('Train shape:', train_df.shape)
print('Test shape: ', test_df.shape)
train_df.head()
```

```python
print(f'Număr exemple train:   {len(train_df)}')
print(f'Număr exemple test:    {len(test_df)}')
print(f'Număr clase (intents): {train_df["label"].nunique()}')
print(f'Lungime medie (cuvinte): {train_df["text"].str.split().str.len().mean():.1f}')

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

label_counts = train_df['label'].value_counts().sort_index()
axes[0].bar(label_counts.index, label_counts.values, color='steelblue', alpha=0.7)
axes[0].axhline(y=label_counts.mean(), color='red', linestyle='--', label=f'Media: {label_counts.mean():.0f}')
axes[0].set_title('Distribuția claselor')
axes[0].set_xlabel('Label')
axes[0].set_ylabel('Număr exemple')
axes[0].legend()

text_lengths = train_df['text'].str.split().str.len()
axes[1].hist(text_lengths, bins=30, color='coral', alpha=0.7, edgecolor='black')
axes[1].axvline(x=text_lengths.mean(), color='red', linestyle='--', label=f'Media: {text_lengths.mean():.1f}')
axes[1].set_title('Lungimea mesajelor (cuvinte)')
axes[1].set_xlabel('Număr cuvinte')
axes[1].legend()

plt.tight_layout()
plt.savefig('eda_plots.png', dpi=150)
plt.show()
print('📊 Grafice salvate!')
```

```python
X = train_df['text']
y = train_df['label']

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

tfidf = TfidfVectorizer(ngram_range=(1, 2), max_features=50000, sublinear_tf=True)
X_train_tfidf = tfidf.fit_transform(X_train)
X_val_tfidf   = tfidf.transform(X_val)

lr_model = LogisticRegression(max_iter=1000, C=5.0, random_state=42)
lr_model.fit(X_train_tfidf, y_train)

val_preds_lr = lr_model.predict(X_val_tfidf)
acc_lr = accuracy_score(y_val, val_preds_lr)
print(f'✅ Baseline Accuracy: {acc_lr:.4f} ({acc_lr*100:.2f}%)')
```

```python
X_test_tfidf  = tfidf.transform(test_df['text'])
test_preds_lr = lr_model.predict(X_test_tfidf)

submission_baseline = pd.DataFrame({
    'id': test_df['id'],
    'label': test_preds_lr
})
submission_baseline.to_csv('submission_baseline.csv', index=False)
print('✅ Salvat: submission_baseline.csv')
submission_baseline.head()
```

```python
MODEL_NAME = 'bert-base-uncased'
NUM_LABELS = train_df['label'].nunique()
MAX_LENGTH = 64

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

def tokenize(examples):
    return tokenizer(
        examples['text'],
        truncation=True,
        max_length=MAX_LENGTH,
        padding=False
    )

train_data = train_df[['text', 'label']].rename(columns={'label': 'labels'})
X_tr, X_vl = train_test_split(train_data, test_size=0.15, random_state=42, stratify=train_data['labels'])

train_dataset = Dataset.from_pandas(X_tr.reset_index(drop=True))
val_dataset   = Dataset.from_pandas(X_vl.reset_index(drop=True))
test_dataset  = Dataset.from_pandas(test_df[['text']].reset_index(drop=True))

train_dataset = train_dataset.map(tokenize, batched=True)
val_dataset   = val_dataset.map(tokenize, batched=True)
test_dataset  = test_dataset.map(tokenize, batched=True)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=NUM_LABELS
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {'accuracy': accuracy_score(labels, predictions)}

training_args = TrainingArguments(
    output_dir='./results_bert',
    num_train_epochs=7,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=64,
    warmup_steps=100,
    weight_decay=0.01,
    learning_rate=2e-5,
    eval_strategy='epoch',
    save_strategy='epoch',
    load_best_model_at_end=True,
    metric_for_best_model='accuracy',
    logging_steps=50,
    fp16=torch.cuda.is_available(),
    report_to='none'
)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

print('🚀 Începe antrenarea BERT...')
trainer.train()
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=NUM_LABELS
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {'accuracy': accuracy_score(labels, predictions)}

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=5,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=64,
    warmup_steps=100,
    weight_decay=0.01,
    learning_rate=3e-5,
    eval_strategy='epoch',
    save_strategy='epoch',
    load_best_model_at_end=True,
    metric_for_best_model='accuracy',
    logging_steps=50,
    fp16=torch.cuda.is_available(),
    report_to='none'
)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

print('🚀 Începe antrenarea...')
trainer.train()
```

```python
predictions = trainer.predict(test_dataset)
test_preds_bert = np.argmax(predictions.predictions, axis=-1)

submission_final = pd.DataFrame({
    'id': test_df['id'],
    'label': test_preds_bert
})
submission_final.to_csv('submission_final.csv', index=False)
print('✅ Salvat: submission_final.csv')
print(f'Total predicții: {len(submission_final)}')
submission_final.head(10)
```
